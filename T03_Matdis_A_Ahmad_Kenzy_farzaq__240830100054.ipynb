```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def harmonic_sum(n):
    return sum(1/i for i in range(1, n+1))

def prove_harmonic_less_than_2():
    test_values = [10, 100, 1000, 10000, 100000]

    print("Bukti: 1 + 1/2 + 1/3 + ... + 1/n < 2")
    print("-" * 40)

    for n in test_values:
        h_n = harmonic_sum(n)
        print(f"n = {n:6d}: H_n = {h_n:.8f} < 2? {h_n < 2}")

    # Bukti bahwa H_n tidak pernah mencapai 2
    print(f"\nBahkan untuk n = 1,000,000:")
    h_million = harmonic_sum(1000000)
    print(f"H_1000000 = {h_million:.8f} < 2")

if __name__ == "__main__":
    prove_harmonic_less_than_2()
```

Semua Hasil "False"

Program membuktikan bahwa H_n TIDAK < 2
Untuk semua n ≥ 10, selalu berlaku H_n > 2
Ini membantah pernyataan soal asli

```python
import matplotlib.pyplot as plt

def harmonic_sum(n):
    return sum(1/i for i in range(1, n+1))

def correct_harmonic_plot():
    # Test perhitungan dulu
    print("Test nilai H_n:")
    for n in [1, 2, 3, 10, 100]:
        h = harmonic_sum(n)
        print(f"H_{n} = {h:.4f}")

    # Data yang benar
    n_values = list(range(1, 101))  # 1 sampai 100
    h_values = [harmonic_sum(n) for n in n_values]

    plt.figure(figsize=(10, 6))
    plt.plot(n_values, h_values, 'b-', linewidth=2, label='H_n = 1 + 1/2 + ... + 1/n')
    plt.axhline(y=2, color='r', linestyle='--', linewidth=2, label='y = 2')

    # Mark beberapa titik
    points = [1, 2, 10, 50, 100]
    for n in points:
        h = harmonic_sum(n)
        plt.plot(n, h, 'ro', markersize=8)
        plt.text(n, h+0.1, f'({n},{h:.2f})', ha='center')

    plt.xlabel('n')
    plt.ylabel('H_n')
    plt.title('Deret Harmonik')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    correct_harmonic_plot()
```

Deret harmonik H_n = 1 + 1/2 + 1/3 + ... + 1/n terus naik
H₁₀ = 2.93, H₁₀₀ = 5.19
Melewati y = 2 pada n ≈ 4

Meskipun divergen, pertumbuhannya sangat lambat sehingga sering "menipu" - terlihat hampir datar padahal terus naik.
